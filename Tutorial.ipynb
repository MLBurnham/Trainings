```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# You can display graphs
n = np.array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)
x = np.linspace(0, 5, 10)
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

```python
import statsmodels.api as sm
from sklearn import datasets
import numpy as np
import pandas as pd
data = datasets.load_boston()
df = pd.DataFrame(data.data, columns=data.feature_names)
target = pd.DataFrame(data.target, columns=["MEDV"])

import statsmodels.api as sm
print(data.DESCR)
```

```python
X = df["RM"]
y = target["MEDV"]
X = sm.add_constant(X)


model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

#### Notable Algorithms
1. Linear Regression: Continuous, fast, widely applicable
2. Logistic Regresion: Discrete, fast, widely applicable
3. Decision Trees and Forests: classification or regression, robust and easy to apply
4. Clustering: Classification, fast or slow depending on the algorithm, Good for unsupervised tasks, image classification
5. Naive Bayes: Classification, very fast so good for NLP, assumes variables are independent which is often unrealistic
6. K Nearest Neighbors: Classification or regression, computationally heavy, good for data sets with a small number of dimensions. e.g. image classification
7. Support Vector Machines (SVM): Classification, effective, complicated
